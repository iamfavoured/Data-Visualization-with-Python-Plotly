## Violin Plots

```python
### A Combination of box plot using a KDE(Kernel Density Estimation) plot

df_tips=px.data.tips()
px.violin(df_tips,y='total_bill',box=True,points='all',height=600)

px.violin(df_tips, y='tip',x='smoker',color='sex', box=True, points='all', hover_data=df_tips.columns,height=600)

fig=go.Figure()
fig.add_trace(go.Violin(x=df_tips['day'][df_tips['smoker']=='Yes'],y=df_tips['total_bill'][df_tips['smoker']=='Yes'], legendgroup='Yes',scalegroup='Yes',
                       name='Yes',side='negative',line_color='blue'))
fig.add_trace(go.Violin(x=df_tips['day'][df_tips['smoker']=='No'],y=df_tips['total_bill'][df_tips['smoker']=='No'], legendgroup='Yes',scalegroup='Yes',
                       name='Yes',side='positive',line_color='red'))
```
